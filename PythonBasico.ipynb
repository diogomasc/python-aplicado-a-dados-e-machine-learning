### Bloco Markdown 2: Tipos Primitivos, Variáveis e I/O

# 2. Variáveis e Tipos de Dados Primitivos
Variáveis são espaços de memória nomeados aos quais atribuímos valores através do operador `=` (recebe). O Python possui tipagem dinâmica, mas os valores pertencem a classes primitivas específicas:
* **Inteiros (`int`):** Números sem casas decimais (ex: 1, 0, -5).
* **Ponto Flutuante (`float`):** Números decimais. Em Python, a separação decimal é feita por ponto (`.`) e não por vírgula.
* **Strings (`str`):** Cadeias de caracteres (texto), encapsuladas por aspas simples ou duplas.
* **Booleanos (`bool`):** Valores lógicos binários (`True` ou `False`).

A função `type()` é utilizada para descobrir a classe de uma variável. Para interagir com o usuário, utilizamos a função `input()`. Como o `input()` retorna sempre uma *string*, funções de conversão (*casting*) como `int()` ou `float()` são necessárias para operações matemáticas.

Para formatar strings misturando texto e variáveis de forma legível, utilizamos as **f-strings** (adicionando um `f` antes das aspas e variáveis entre `{}`).

```python
### Bloco de Código 2: Manipulação de Variáveisz

# Declaração de Variáveis
idade = 30           # int
altura = 1.75        # float
nome = 'Maria'       # str
estudante = True     # bool

# Verificando os tipos
print(type(altura))  # Retornará <class 'float'>

# Interação e Conversão (Casting)
# O input é envolvido por int() para garantir que a entrada seja lida como número inteiro.
# idade_usuario = int(input("Digite sua idade: "))

# F-strings para formatação limpa e moderna
print(f"Olá {nome}, você tem {idade} anos e mede {altura}m.")

# Tamanho de uma string e fatiamento (slicing - sempre começa no índice 0)
print(f"O nome {nome} tem {len(nome)} letras.")
print(f"As 3 primeiras letras são: {nome[0:3]}") # Retorna 'Mar'
```

### Bloco Markdown 3: Operadores e Controle de Fluxoz

# 3. Operadores e Controle de Fluxo
A lógica de programação depende de avaliar condições. O Python suporta operadores aritméticos (`+`, `-`, `*`, `/`, `**` para potência, `%` para resto da divisão) e operadores relacionais (`==`, `!=`, `>`, `<`, `>=`, `<=`). 

Comparações lógicas compostas são feitas com `and` (conjunção), `or` (disjunção) e `not` (negação).

**Condicionais (`if`, `elif`, `else`):**
O Python utiliza a **indentação** (espaçamentos à direita) no lugar de chaves `{}` ou palavras-chave (como `begin/end`) para delimitar blocos de código. Uma condicional determina qual bloco de código será executado com base em um teste lógico.

```python
### Bloco de Código 3: Condicionais e Operadores

numero = 8

# Verificando se um número é par usando o operador Módulo (%)
# O bloco de código pertencente à condicional deve estar indentado.
if numero % 2 == 0:
    print(f"O número {numero} é PAR.")
elif numero % 3 == 0:
    print(f"O número {numero} é divisível por 3.")
else:
    print(f"O número {numero} é ÍMPAR e não é divisível por 3.")
```

### Bloco Markdown 4: Estruturas de Repetição (Loops)

# 4. Estruturas de Repetição
Laços de repetição automatizam a execução de blocos de código.
* **`for`:** Ideal para percorrer coleções ou intervalos conhecidos. Comumente associado à função `range(start, stop, step)`, lembrando que o Python sempre exclui o último valor (o limite superior é aberto).
* **`while`:** Executa um bloco enquanto uma condição for verdadeira (`True`). Exige que a variável de controle seja atualizada manualmente (incremento como `+= 1`); caso contrário, entraremos em um laço infinito (*loop infinito*).

```python
### Bloco de Código 4: Loops

print("Exemplo de FOR:")
# range(0, 10, 2) vai de 0 até 9, pulando de 2 em 2
for i in range(0, 10, 2):
    print(i, end=" ") # end=" " evita pular linha

print("\n\nExemplo de WHILE:")
contador = 0
while contador < 5:
    print(contador, end=" ")
    contador += 1 # Incremento manual vital para não criar loop infinito
```

### Bloco Markdown 5: Coleções (Listas, Tuplas e Dicionários)

# 5. Coleções (Estruturas de Dados)
Para agrupar múltiplos valores em uma única variável, usamos coleções:
* **Listas (`list`):** Definidas por colchetes `[]`. São mutáveis e ordenadas. Permitem adição (`append()`), remoção (`pop()`, `remove()`), fatiamento e operações como multiplicação para duplicação. Possuem a sintaxe avançada de *List Comprehension* para aplicar loops em uma única linha.
* **Tuplas (`tuple`):** Definidas por parênteses `()`. Semelhantes às listas, porém são **imutáveis**. Após criadas, não suportam alteração, adição ou remoção de itens.
* **Dicionários (`dict`):** Definidos por chaves `{}`. Mapeiam pares de `chave: valor` (como vetores associativos). As chaves devem ser rigorosamente únicas. Úteis para recuperação rápida de dados baseada em rótulos, acessados por `.keys()`, `.values()` e `.items()`.

```python
### Bloco de Código 5: Coleções

# LISTAS
lista_quadrados = []
lista_quadrados.append(25) # Adiciona ao final
# lista_quadrados.pop(0)     # (Comentado para não dar erro se a lista estiver vazia antes das operações)

# List Comprehension (Jeito acadêmico e pythônico de aplicar loops em listas)
lista_dobro = [valor * 2 for valor in lista_quadrados]
print("List Comprehension:", lista_dobro)

# TUPLAS
minha_tupla = (10, 20, 30)
# minha_tupla = 15 # Isso geraria um TypeError, pois tuplas são imutáveis

# DICIONÁRIOS
aluno = {
    "nome": "Ana",
    "idade": 22,
    "curso": "Engenharia"
}

# Iterando simultaneamente sobre chave e valor
for chave, valor in aluno.items():
    print(f"{chave.capitalize()}: {valor}")
```

### Bloco Markdown 6: Funções

# 6. Funções
As funções encapsulam blocos de código concebidos para realizar tarefas específicas e parametrizadas. 
* **Reusabilidade:** Evitam a repetição de código (como uma fórmula de C° para F°) em múltiplas telas.
* **Definição:** Declaradas com a palavra-reservada `def`.
* **Parâmetros e Retorno:** Podem receber parâmetros (inclusive com valores *default* opcionais) e retornar valores processados através do comando `return`, que encerra a execução da função imediatamente.

```python
### Bloco de Código 6: Funções

# Função com valor default (padronizado). 
# Se celsius não for fornecido, assumirá 30.
def converter_celsius_fahrenheit(celsius=30):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Chamando a função
temp_f = converter_celsius_fahrenheit(25)
print(f"25ºC equivalem a {temp_f}ºF.")

# Chamando com o valor padrão
temp_padrao = converter_celsius_fahrenheit()
print(f"Valor padrão de 30ºC equivale a {temp_padrao}ºF.")
```
